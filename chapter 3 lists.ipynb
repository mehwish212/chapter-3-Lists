```python
#try It yourself 3-8. Seeing the World: 

places = ['paris','malaysia','saudi arabia','australia','london']
print('original order')
print(places)

print('\nalphabatical order temporarily')
print(sorted(places))

print('\nlist stil in original order')
print(places)

print("\nreverse alphabetical order temporarily")
print(sorted(places ,reverse=True))
print(places)

print('\nlist stil in original order')
print(places)

print("\nchange the order of your list .")
places.reverse()
print(places)

print("\nusing reverse again to change the order of your list to its origial order .")
places.reverse()
print(places)

print("  \nlist is stored in alphabetical order permanently .")
places.sort()
print(places)

print("\nlist order has been changed permanently")
print(places)

print("\nlist is stored in reverse alphabetical order permanently .")
places.sort(reverse = True)
print(places)

print("\nlist has been changed reverse alphabetical order permanently")
print(places)
```
